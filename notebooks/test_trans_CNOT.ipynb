```python
import sys
import os

# Add project root to path
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
    sys.path.insert(0, project_root)

import stim

from experiments.CNOT_trans import CNOTTransExperiment
```

```python
# -----------------------------
# Config
# -----------------------------
distance = 3
offset_target = (6, 0)  # target patch offset relative to control patch

initial_basis_control = "X"   # "X" or "Z"
initial_basis_target = "Z"    # "X" or "Z"

measure_basis_control = "Z"   # "X" or "Z"
measure_basis_target = "Z"    # "X" or "Z"

rounds_before = 2
rounds_after = 2
if_detector = True

# -----------------------------
# Create and run the experiment
# -----------------------------
experiment = CNOTTransExperiment(
    distance=distance,
    offset_target=offset_target,
    initial_basis_control=initial_basis_control,
    initial_basis_target=initial_basis_target,
    measure_basis_control=measure_basis_control,
    measure_basis_target=measure_basis_target,
    rounds_before=rounds_before,
    rounds_after=rounds_after,
    if_detector=if_detector
)

circuit = experiment.build()

print("--------------------------------")
print("Number of logical observables:", circuit.num_observables)

# Visualization
circuit.diagram("detslice-with-ops-svg")
```
