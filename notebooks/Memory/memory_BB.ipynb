# Memory Experiment — Bivariate Bicycle (BB) Codes

High-rate LDPC codes from Bravyi et al. 2024.  
Key codes: [[72,12,6]], [[108,8,10]], [[144,12,12]] ("Gross code").

```python
import sys
from pathlib import Path
import numpy as np

ROOT = Path("../..").resolve()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from lightstim.ir.qec_system import QECSystem
from lightstim.protocols.memory import MemoryExperiment
from lightstim.qec_code.BB_code import BBCode, BBCodeExtractionBlock
```

```python
from lightstim.noise.config import NoiseConfig

p = 1e-3
noise_params = NoiseConfig(p_idle=p, p_1q=p, p_2q=p, p_meas=p, p_reset=p)
```

## BB code configurations

Each BB code is parameterized by `l, m, A, B` matrices over `Z_l × Z_m`.

```python
BB_CONFIGS = {
    "bb_72_12_6":   {"l": 6,  "m": 6,  "A": [[3,0],[0,1],[0,2]], "B": [[0,3],[1,0],[2,0]]},
    "bb_108_8_10":  {"l": 9,  "m": 6,  "A": [[3,0],[0,1],[0,2]], "B": [[0,3],[1,0],[2,0]]},
    "bb_144_12_12": {"l": 12, "m": 6,  "A": [[3,0],[0,1],[0,2]], "B": [[0,3],[1,0],[2,0]]},
}
```

## [[72,12,6]] — Smallest BB Code

```python
cfg = BB_CONFIGS["bb_72_12_6"]
code = BBCode(l=cfg["l"], m=cfg["m"], A=cfg["A"], B=cfg["B"])
system = QECSystem()
system.add_patch(code, name="bb_72_12_6")

exp = MemoryExperiment(
    qec_system=system,
    extraction_block_class=BBCodeExtractionBlock,
    rounds=3,
    noise_params=noise_params,
    noise_model="circuit_level",
    basis="Z",
)
circuit_72 = exp.build()
print(f"Qubits: {circuit_72.num_qubits}  Detectors: {circuit_72.num_detectors}  k={code.num_logicals}")
circuit_72
```

```python
cfg = BB_CONFIGS["bb_144_12_12"]
code = BBCode(l=cfg["l"], m=cfg["m"], A=cfg["A"], B=cfg["B"])
system = QECSystem()
system.add_patch(code, name="bb_144_12_12")

exp = MemoryExperiment(
    qec_system=system,
    extraction_block_class=BBCodeExtractionBlock,
    rounds=3,
    noise_params=noise_params,
    noise_model="circuit_level",
    basis="Z",
)
circuit_144 = exp.build()
print(f"Qubits: {circuit_144.num_qubits}  Detectors: {circuit_144.num_detectors}  k={code.num_logicals}")
# Diagram omitted — very large circuit; use circuit_144 to inspect Stim instructions
```
