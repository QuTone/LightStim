# Bell State Teleportation Circuits

Three implementations of logical Bell teleportation on the unrotated surface code:

- **(a) TG-based**: Transversal CNOT gates
- **(b) ZZ-LS-based**: Two sequential ZZ lattice surgery measurements
- **(c) XX-LS-based**: Two sequential XX lattice surgery measurements

Protocol: teleport `|П€вџ©_L` from patch1 to patch3 via an entangled patch2.

````python
from pathlib import Path
import sys
ROOT = Path('../..').resolve()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from lightstim.protocols.bell_teleportation import (
    BellTeleportTG,
    BellTeleportZZLS,
    BellTeleportXXLS,
)
````

---
## (a) Transversal-Gate-Based Bell Teleportation

```
patch1 |П€вџ©_L в”Ђв”Ђв”Ђв”Ђв”Ђв”ЂSEв”Ђв”Ђв—Џв”Ђв”ЂSEв”Ђв”Ђв”Ђв”Ђ Meas X в†’ a
                        в”‚
patch2 |+вџ©_L в”Ђв”Ђв”Ђв—Џв”Ђв”ЂSEв”Ђв”ЂвЉ•в”Ђв”ЂSEв”Ђв”Ђв”Ђв”Ђ Meas Z в†’ b
                в”‚
patch3 |0вџ©_L в”Ђв”Ђв”ЂвЉ•в”Ђв”ЂSEв”Ђв”Ђв”Ђв”Ђв”Ђв”ЂSEв”Ђв”Ђв”Ђв”Ђ teleport_state^a Z^b в†’ |П€вџ©_L
```

Layout: three patches placed horizontally.

````python
import contextlib, io

distance = 3
teleport_state = 'Z'   # 'Z' or 'X'

exp_tg = BellTeleportTG(
    distance=distance,
    rounds_pre= 1,
    rounds_mid= 1,
    rounds_post= 1,
    teleport_state=teleport_state,
    if_detector=True,
)
with contextlib.redirect_stdout(io.StringIO()):
    circuit_tg = exp_tg.build()

print(f'TG Bell Teleportation  (d={distance}, state={teleport_state})')
print(f'  num_qubits:      {circuit_tg.num_qubits}')
print(f'  num_detectors:   {circuit_tg.num_detectors}')
print(f'  num_observables: {circuit_tg.num_observables}')
````
