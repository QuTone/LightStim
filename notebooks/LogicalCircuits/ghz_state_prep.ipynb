```python
import sys
from pathlib import Path

ROOT = Path('../..').resolve()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import stim

from lightstim.protocols.ghz import GHZExperiment
from lightstim.noise.config import NoiseConfig
```

```python
d = 3

exp = GHZExperiment(
    distance=d,
    offset_patch2=(4*(2*d-1)-2, 0),
    offset_patch3=(0, 4*(2*d-1)-2),
    initial_basis_patch1="X",   # |+>
    initial_basis_patch2="Z",   # |0>
    initial_basis_patch3="Z",   # |0>
    measure_basis_patch1="Z",
    measure_basis_patch2="Z",
    measure_basis_patch3="Z",
    rounds_before=d,
    rounds_after=1,
)
circuit = exp.build()

print(f"d={d} | qubits={circuit.num_qubits}, detectors={circuit.num_detectors}, observables={circuit.num_observables}")
circuit.diagram("detslice-with-ops-svg")
```

```python
# Test case: Measure in Z, Z, Z basis
# This should show correlations in the logical observables
print("Testing Z, Z, Z measurement...")
print("Expected: GHZ state |000> + |111> should show correlations")
circuit.detector_error_model()
```
